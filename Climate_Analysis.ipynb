```python
import itertools
import re
import nltk
from tqdm import tqdm

# Define a function to extract hashtags from a tweet
def extract_hashtags(tweet):
    return re.findall(r"#(\w+)", tweet)

# Create an empty dictionary to store the top 15 hashtags for each sentiment
top15 = {}

# Group the tweets by sentiment
by_sentiment = tweets.groupby("sentiment")

# Iterate over each sentiment group and extract the hashtags
for sentiment, group in tqdm(by_sentiment):
    hashtags = group["message"].apply(extract_hashtags)
    hashtags = list(itertools.chain(*hashtags))
    hashtags = [ht.lower() for ht in hashtags]
    
    # Count the frequency of each hashtag using nltk
    frequency = nltk.FreqDist(hashtags)
    
    # Create a DataFrame with the top 15 hashtags for this sentiment
    top15_hashtags = pd.DataFrame({
        "hashtags": list(frequency.keys()),
        "counts": list(frequency.values()),
    }).nlargest(15, columns=["counts"]).reset_index(drop=True)
    
    # Add the top 15 hashtags to the dictionary
    top15[sentiment] = top15_hashtags

# Concatenate the top 15 hashtags for each sentiment into a single DataFrame and display the results
top15_df = pd.concat(top15, axis=1)
display(top15_df.head(n=15))
```

```python
#data cleaning using regex
def clean_tweet(tweet):
# Convert the tweet to lowercase
    tweet = tweet.lower()
    
    
# Define a list of patterns to remove from the tweet
    to_del = [
        r"@[\w]*",  #remove account mentions
        r"http(s?):\/\/.*\/\w*",  #remove URLs
        r"#\w*",  #remove hashtags
        r"\d+",  # delete numeric values
        r"U+FFFD",  # remove the "character note present" diamond
    ]

# Loop through the patterns and remove them from the tweet
    for key in to_del:
        tweet = re.sub(key, "", tweet)
    
   # Remove punctuation and special characters
    tweet = re.sub(r"[,.;':@#?!\&/$]+\ *", " ", tweet)
   # Remove excess white space
    tweet = re.sub(r"\s\s+", " ", tweet)
    
    return tweet.lstrip(" ")
```

```python
def lemmatizer(df):
    df["length"] = df["message"].str.len()
    df["tokenized"] = df["message"].apply(word_tokenize)
    df["parts-of-speech"] = df["tokenized"].apply(nltk.tag.pos_tag)
    
    def string_to_wordnet(tag):
        conversion = {"J": wordnet.ADJ, "V": wordnet.VERB, "N": wordnet.NOUN, "R": wordnet.ADV}
        try:
            return conversion[tag[0].upper()]
        except KeyError:
            return wordnet.NOUN
    
    wnl = WordNetLemmatizer()
    df["parts-of-speech"] = df["parts-of-speech"].apply(
        lambda tokens: [(word, string_to_wordnet(tag)) for word, tag in tokens]
    )
    df["lemmatized"] = df["parts-of-speech"].apply(
        lambda tokens: [wnl.lemmatize(word, tag) for word, tag in tokens]
    )
    df["lemmatized"] = df["lemmatized"].apply(lambda tokens: " ".join(map(str, tokens)))
    
    return df
```

```python
import spacy

# Load the English model in Spacy
spacy_en = spacy.load('en_core_web_sm')
```

```python
# Define function for crude entity extraction
def crude_entities(tweet):

#Tokenize tweets into words using the Spacy model
    as_words = tweet.apply(spacy_en)
    
#Define function to filter words by label
    def by_label(words, label):
        filtered = [word.text for word in words.ents if word.label_ == label]
        return filtered
    
#Define function to get top entities by label
    def get_top(label, n=10):
        thing = as_words.apply(lambda x: by_label(x, label))
        flattened = itertools.chain(*thing.values)
        
        counter = Counter(flattened)
        topN = counter.most_common(n)
        
        topN_things = [thing for thing, _ in topN]
        
        return thing

#Create a DataFrame to store the top entities
    entities = pd.DataFrame()
    entities["people"] = get_top("PERSON", n=10)
    entities["geopolitics"] = get_top("GPE", n=10)
    entities["organizations"] = get_top("ORG")
    
    return entities
```

```python
# Create an empty dictionary to store the entities by sentiment
entities = {}

# Group the tweets by sentiment using pandas
by_sentiment = tweets.groupby("sentiment")

# Loop through each sentiment group and extract the entities using the crude_entities function
for sentiment, group in tqdm(by_sentiment):
    entities[sentiment] = crude_entities(group["lemmatized"])
    
# Combine the entities from each sentiment group into a single DataFrame for display
display(pd.concat(entities, axis=1).head(n=100))
```

```python
X = tweets["message"]
Y = tweets["sentiment"]
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training, validation, and testing sets using the train_test_split function
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=40)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.20, random_state=40)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create a CountVectorizer object to convert the text data into bag-of-words vectors
tfidf = CountVectorizer()

# Use the fit_transform method to vectorize the training data and transform the validation and testing data
X_train = tfidf.fit_transform(X_train)
X_test = tfidf.transform(X_test)
X_valid = tfidf.transform(X_valid)
```

```python
def train_Model(model, train_data, train_labels, test_data):

    model.fit(train_data, train_labels)
    
    return model, model.predict(test_data)
```

```python
def evaluate_Model(model, preds, test_labels):
    print(metrics.classification_report(test_labels, preds))
    
    cm = confusion_matrix(test_labels, preds)
    cm_normd = cm / cm.sum(axis=1).reshape(-1, 1)
    
    sns.heatmap(cm_normd, cmap="YlGnBu", xticklabels=model.classes_,
        yticklabels=model.classes_,
        vmin=0.,
        vmax=1.,
        annot_kws={"size": 10})
    
    plt.title(f"{model.__class__.__name__} Classification")
    plt.ylabel("Ground-truth labels")
    plt.xlabel("Predicted labels")
    plt.plot()
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
svm_linear = LinearSVC(penalty='l2', loss='squared_hinge', class_weight="balanced")
svm_linear.fit(X_train, y_train)
preds = svm_linear.predict(X_valid)
evaluate_Model(svm_linear, preds, y_valid)
```

```python
KNN = KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
KNN, preds = train_Model(KNN, X_train, y_train, X_valid)
evaluate_Model(KNN, preds, y_valid)
```

```python
# Instantiate Multinomial Naive Bayes classifier
nb = MultinomialNB()

# Train the Multinomial Naive Bayes classifier on the training data and validate on the validation data
nb, preds = train_Model(nb, X_train, y_train, X_valid)

# Evaluate the performance of the Multinomial Naive Bayes classifier on the validation data
evaluate_Model(nb, preds, y_valid)
```

```python
# Instantiate Random Forest classifier with specified hyperparameters
rf = RandomForestClassifier(max_depth=5, n_estimators=100)

# Train the Random Forest classifier on the training data and validate on the validation data
rf, preds = train_Model(rf, X_train, y_train, X_valid)

# Evaluate the performance of the Random Forest classifier on the validation data
evaluate_Model(rf, preds, y_valid)
```

```python
# Instantiate Logistic Regression classifier with specified hyperparameters
LR = LogisticRegression(C=1, class_weight="balanced", max_iter=1000)

# Train the Logistic Regression classifier on the training data and validate on the validation data
LR, preds = train_Model(LR, X_train, y_train, X_valid)

# Evaluate the performance of the Logistic Regression classifier on the validation data
evaluate_Model(LR, preds, y_valid)
```

It is observed that, the Logistic Regression model gives the highest accuracy of about 70% compared to other models.
